```python

# Approach 2
Y= data["left"]
X = data.drop(['left'], axis=1)
```

```python
# Train and test split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=62, stratify=Y)
x_train.shape
x_test.shape
y_train.shape
y_test.shape
```
